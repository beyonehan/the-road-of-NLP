```python
import pandas as pd 
import numpy as np
```

```python
news = pd.read_csv('sqlResult_1558435.csv',encoding='gb18030')
```

```python
len(news)
```

```python
news.dropna(inplace=True)
```

```python
news.info()
```

```python
news.replace({'source':r'^(新华社)'},{'source':1},regex=True, inplace=True)
```

```python
news.loc[news['source'] != 1 ,'source'] = 0
```

```python
news.head(300)
```

```python
with open('./stop_words.utf8') as f:
    
    stop_words =  [l.strip() for l in f]
```

```python
import jieba
```

```python
def cut(string) : return [word for word in list(jieba.cut(string)) if word in stop_words] 
```

```python
all_content = news['content']
```

```python
all_content = [''.join(cut(s)) for s in all_content ]
```

```python
import pickle
```

```python
with open('news_content_words.pkl','wb') as f:
    pickle.dump(all_content,f)
    
```

```python
with open('news_content_words.pkl','rb') as f:
    all_content = pickle.load(f)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer(max_features=50)
```

```python
X = vectorizer.fit_transform(all_content)
```

```python
print( vectorizer.get_feature_names())
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
y = pd.to_numeric(news['source']).values
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.7,random_state = 0)
```

```python
X_train.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(X_train,y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
from sklearn import metrics
```

```python
print ('accuracy:',metrics.accuracy_score(y_test,y_pred))
```
